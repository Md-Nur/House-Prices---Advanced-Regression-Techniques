# Missing Value

```python
catFeatureWithNull = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>0 and dataset[feature].dtype == 'O']
for i in catFeatureWithNull:
    print(f"{i} : {np.round(np.mean(dataset[i].isnull())*100,2)} %")
```

```python
dataset[catFeatureWithNull] = dataset[catFeatureWithNull].fillna('Missing')
```

```python
dataset[catFeatureWithNull].isnull().sum()
```

```python
dataset.head()
```

```python
numFeatureWithNull = [f for f in dataset.columns if dataset[f].dtype !='O' and dataset[f].isnull().sum()>0]
numFeatureWithNull
```

```python
for f in numFeatureWithNull:
    medianVal = np.nanmedian(dataset[f])
    dataset[f+'nan']=np.where(dataset[f].isnull(),0,1)
    dataset[f].fillna(medianVal,inplace=True)
dataset[numFeatureWithNull].isnull().sum()
```

```python
dataset.head(10)
```

```python
timeFeature = np.array(['YearBuilt', 'YearRemodAdd', 'GarageYrBlt'])
for t in timeFeature:
    dataset[t] = dataset['YrSold'] - dataset[t]
```

```python
dataset[timeFeature]
```

# Numeric Value

```python
dataset.head()
```

```python
numFeature = np.array(['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice'])
for nf in numFeature:
    dataset[nf] = np.log(dataset[nf])
```

```python
dataset.head()
```

```python
catFeature = [f for f in dataset.columns if dataset[f].dtype=='O']
catFeature
```

```python
for cf in catFeature:
    temp = dataset.groupby(f)['SalePrice'].count()/len(dataset)
    tempDf = temp[temp>0.01].index
    dataset[cf] = np.where(dataset[cf].isin(tempDf),dataset[cf],"RareVal")
```
