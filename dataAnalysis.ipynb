# Handle Null Values

```python
featureWithNull = [f for f in dataset.columns if dataset[f].isnull().any()]
featureWithNull.remove('SalePrice')
len(featureWithNull)
# featureWithNull
```

```python
for f in featureWithNull:
    print(f,dataset[f].isnull().mean())
    if dataset[f].isnull().mean()>0.5:
        dataset.drop(f,axis=1,inplace=True)
        featureWithNull.remove(f)
```

```python
dataset.head()
```

```python
missingNum = [f for f in featureWithNull if dataset[f].dtype!='O']
len(missingNum)
```

```python
data = dataset.copy()
for mn in missingNum:
    if 0 not in data[mn].unique():
        data[mn]=np.log(data[mn])
        data.boxplot(column=mn)
        plt.show()
```

```python

for nm in missingNum:
    medianVal = np.nanmedian(dataset[nm])
    print(nm,medianVal)
    dataset[nm].fillna(medianVal,inplace=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
missingCatVal = []

for f in featureWithNull:
    if f not in missingNum:
        missingCatVal.append(f)
        
```

```python
len(missingCatVal)
```

```python
missingCatVal
```

```python
dataset[missingCatVal]=dataset[missingCatVal].fillna('Missing')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.heatmap(dataset.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

## Handle Year

```python
timeFeature = np.array(['YearBuilt', 'YearRemodAdd', 'GarageYrBlt'])
for t in timeFeature:
    dataset[t] = dataset['YrSold'] - dataset[t]
dataset[timeFeature]
```

# Log Transform

```python
for mn in missingNum:
    ax = sns.displot(dataset[mn])
    print(mn)
    plt.show()
```

```python
checkLog = ['LotFrontage','MasVnrArea','BsmtFinSF1','BsmtFinSF2','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','GarageYrBlt']
dataLog = dataset.copy()
for f in checkLog:
    logVal = np.log(dataLog[f])
    dataLog[f] = logVal
    ax1 = sns.displot(dataLog[f])
    print(f)
    plt.show()
```

```python
dataset.head()
```

```python
applyLog = ['LotFrontage','BsmtFinSF1','TotalBsmtSF','GarageYrBlt']
for f in applyLog:
    logVal = []
    for i in dataset[f]:
        if i!=0:
            logVal.append(np.log(i))
        else:
            logVal.append(i)
            
    dataset[f] = logVal
    print(f)
```

```python
dataset.head()
```

# Create Dummy Var

```python
catVal = [f for f in dataset.columns if dataset[f].dtype=='O' and len(dataset[f].unique())<10]
len(catVal)
```

```python
for f in dataset.columns:
    if dataset[f].dtype=='O' and len(dataset[f].unique())>=10:
        dataset.drop(f,axis=1,inplace=True)
```

```python
dataset.head()
```

```python
for f in catVal:
    df1 = pd.get_dummies(dataset[f],drop_first=True)
    dataset = pd.concat([dataset,df1],axis=1)
    dataset.drop(f,axis=1,inplace=True)
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt','log2']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 1000,10)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10,14]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4,6,8]
# Create the random grid
criterion = ['squared_error','absolute_error','poisson']
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
              'criterion':criterion}
print(random_grid)
```
